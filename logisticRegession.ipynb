```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1/(1+np.exp(-x))
# load data from 
data = pd.read_csv("logistic_dataset.csv").values
```

```python
N,d = data.shape
x = data[:,0:d-1].reshape(-1,d-1)
y = data[:,2].reshape(-1,1)
print(x)

print(y)
```

```python
# ve data bang scatter
x_cho_vay = x[y[:,0]==1]
x_tu_choi = x[y[:,0]==0]

plt.scatter(x_cho_vay[:,0],x_cho_vay[:,1],c='red',edgecolors='none',s=30,label='cho vay')
plt.scatter(x_tu_choi[:,0],x_tu_choi[:,1],c='blue',edgecolors='none',s=30,label='tu choi')
plt.legend(loc=1)
plt.xlabel('muc luong (trieu )')
plt.xlabel('kinh nghiem (nam)')
plt.show()
```

```python
# them cot 1 vao du lieu x 
x = np.hstack((np.ones((N,1)),x))
w = np.array([0.,0.1,0.1]).reshape(-1,1)
```

```python
numOfIteration = 1000
cost = np.zeros((numOfIteration,1))
learning_rate = 0.01
```

```python
for i in range(1,numOfIteration):
    # tinh gia tri du doan 
    y_predict = sigmoid(np.dot(x,w))
    cost[i]= -np.sum(np.multiply(y,np.log(y_predict))+np.multiply(1-y,np.log(1-y_predict)))
    # gradient descent
    w = w - learning_rate* np.dot(x.T,y_predict-y)
    print(cost[i])

# ve duong phan cach 
t = 0.5
plt.plot((4, 10),(-(w[0]+4*w[1]+ np.log(1/t-1))/w[2], -(w[0] + 10*w[1]+ np.log(1/t-1))/w[2]), 'g')
plt.show()
```
